```python
import os
from PIL import Image
from matplotlib.pyplot import figure, imshow, axis
from matplotlib.image import imread
import matplotlib.pyplot as plt
%matplotlib inline
import images_ops
from images_ops import open, filter, blur, flip, denoise, sharpen
```

```python
def showImages(img,after = None):
    if after:
        f,ax = plt.subplots(1,2,figsize=(12, 6))
        ax[0].imshow(img.convert("RGB"))  
        ax[1].imshow(after.convert("RGB"))
        ax[0].axis('off')
        ax[1].axis('off')
        plt.show()
    else:
        plt.axis('off')
        plt.imshow(img.convert("RGB"))
```
